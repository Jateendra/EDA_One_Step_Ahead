# Expression System

```python
Dont waste memory or time with feature engineering , we (lazily) transform your data when needed  .
```

```python
%%time
vaex_df['division_col13']=vaex_df.col1/vaex_df.col3
```

```python
vaex_df['division_col13']
```

```python
%%time
vaex_df['multiplication_col13']=vaex_df.col1*vaex_df.col3
```

```python
vaex_df['multiplication_col13']
```

# Out-of-core DataFrame

```python
#Filtering and evaluating  expressions will not waste memory by making copies , the data is kep untouched on disk , 
#and will be streamed only when needed . Delay the time before you need a cluster .
```

```python
vaex_df[vaex_df.col2>70]
```

```python
dff = vaex_df[vaex_df.col2>70] # Here filtering will not amek a memory copy
```

```python
# All the algrithms work  out of core , the limit is the size of your harddriver
dff.col2.mean(progress='widget')
```

```python
dff.col2.min(progress='widget')
```

# Fast Groupby / Aggregations

Vaex implements parallelized , highly peformant groupby operations,especially when using categories ( >1 billion/second).

```python
%%time
vaex_df_group=vaex_df.groupby(vaex_df.col1,agg=vaex.agg.mean(vaex_df.col3))
vaex_df_group
```

```python
%%time
vaex_df.groupby(vaex_df.col1,agg='count')
```
